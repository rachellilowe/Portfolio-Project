```python
#import libraries
import pandas as pd
import numpy as np
```

```python
# read the file
hotels = pd.read_csv('C:\\Users\\lolic\\Desktop\\Data science\\dataset\\Kaggle\\hotel booking\\hotel_bookings.csv')
```

```python
#explore the file and load the first 5 rows of data
hotels.head()
```

```python
# check out how many rows of this data
len(hotels)
```

```python
# find out if there is any missing data
hotels.isnull().sum()
```

```python
#children has 4 missing,country 488 missing,agent 16340 missing and company has the most missing 112593,drop 'company'column to clean the data
hotels = hotels.drop('company',axis=1)
hotels
```

```python
# top 10 country in the dataset
hotels['country'].value_counts()[:10]
```

```python
#The mean ADR(average daily rate) of all the hotel stays in the dataset
np.round(hotels['adr'].mean(),decimals = 2)
```

```python
#The average(mean) number of nights for a stay across the entire dataset
#Total stay nights include 'stay in week nights' and 'stays in weekend nights'
hotels['total_stay_nights'] = hotels['stays_in_week_nights'] + hotels['stays_in_weekend_nights']
round(hotels['total_stay_nights'].mean(),2)
```

```python
# The average(mean) total cost for a stay
hotels['total_spend'] = hotels['adr']*hotels['total_stay_nights']
round(hotels['total_spend'].mean(),2)
```

```python
#The highest paid ADR(average daily rate)
hotels[['adr','total_stay_nights','reservation_status']].max()
```

```python
# The percentage of 'repeat guests'
round(100 * sum(hotels['is_repeated_guest']==1) / len(hotels),2)
```

```python
#Top 5 bookings with the most number of children and babies
hotels['total_kids'] = hotels['children'] + hotels['babies']
hotels.nlargest(5,'total_kids')[['hotel','adr','reservation_status','total_stay_nights','total_kids','babies','children']]
```
